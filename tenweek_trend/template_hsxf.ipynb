```python
import pandas as pd 
from datetime import datetime 
```

```python
df = pd.read_excel(".\\data\\hsxf.xlsx")
df.head()
```

```python
datetime.now().weekday()


weekday_dict = {
    0 : "Monday",
    1 : "Tuesday",
    2 : "Wednesday",
    3 : "Thursday",
    4 : "Friday",
    5 : "Saturday",
    6 : "Sunday",
    
}
```

```python
df.SHPD_DTT = df.SHPD_DTT.apply(pd.to_datetime)
df.IOD_REPORT_DTT = df.IOD_REPORT_DTT.apply(pd.to_datetime)
df.ETA = df.ETA.apply(pd.to_datetime)
df.DELIVERY_DTT = df.DELIVERY_DTT.apply(pd.to_datetime)
```

```python
df['SHPD_DTT_weekday'] = df.SHPD_DTT.apply(lambda x : weekday_dict[x.weekday()])
df['ETA_weekday'] = df.ETA.apply(lambda x : weekday_dict[x.weekday()])
df['DELIVERY_DTT_weekday'] = df.DELIVERY_DTT.apply(lambda x : weekday_dict[x.weekday()])
df['IOD_REPORT_DTT_weekday'] = df.IOD_REPORT_DTT.apply(lambda x : weekday_dict[x.weekday()])
```

```python
df.to_csv("hsxf_table.csv",index = False)
```
